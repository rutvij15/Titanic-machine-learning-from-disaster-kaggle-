```python
train['Age'].fillna(train['Age'].median() ,inplace=True)
test['Age'].fillna(test['Age'].median(), inplace=True)

print(train['Age'].isnull().sum())
print(test['Age'].isnull().sum())
```

```python
train.drop("Cabin" , axis=1)
test.drop("Cabin", axis=1)
```

```python
train['Embarked'].fillna(train['Embarked'].mode()[0] , inplace=True)
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
test['Fare'].fillna(test['Fare'].median() , inplace=True)
print(test.isnull().sum())
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python

print("Train Sex unique values: ", train['Sex'].unique())
print("Test Sex unique values: ", test['Sex'].unique())
```

```python

print("Train missing values in 'Sex':", train['Sex'].isnull().sum())
print("Test missing values in 'Sex':", test['Sex'].isnull().sum())
```

```python

train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})
test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})


print(train['Sex'].head())
print(test['Sex'].head())
```

```python

train = train.drop(columns=['Name', 'Ticket', 'Cabin'])
test = test.drop(columns=['Name', 'Ticket', 'Cabin'])  


print(train.head())
print(test.head())
```

```python

train = pd.get_dummies(train, columns=['Embarked'], drop_first=True)  
test = pd.get_dummies(test, columns=['Embarked'], drop_first=True)


print(train.head())
print(test.head())
```

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


x_train = train.drop(columns=['Survived'])  
y_train = train['Survived'] 
x_test = test


x_train_split, x_val_split, y_train_split, y_val_split = train_test_split(
    x_train, y_train, test_size=0.2, random_state=42
)


model = RandomForestClassifier(random_state=42)


model.fit(x_train_split, y_train_split)


y_val_pred = model.predict(x_val_split)


val_accuracy = accuracy_score(y_val_split, y_val_pred)
print(f'Accuracy on validation set: {val_accuracy}')


y_test_pred = model.predict(x_test)


```

```python

submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],  
    'Survived': y_test_pred  
})


submission.to_csv('submission.csv', index=False)
```
